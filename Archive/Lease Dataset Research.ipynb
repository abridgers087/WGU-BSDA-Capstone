```python
import kagglehub
import os
import json
```

```python
# Download latest version
path = kagglehub.dataset_download("pranjalverma08/sec-edgar-annual-financial-filings-2021")

print("Path to dataset files:", path)
extract_path = os.path.join(path, "extracted")
```

```python
sample_json = os.path.join(extract_path, "1001601_10K_2020_0001493152-21-008913.json")

with open(sample_json, "r", encoding="utf-8") as f:
    data = json.load(f)

print(type(data))
print(data.keys())  
```

```python
def classify_item6(text):
    if not text:
        return "empty"
    t = text.lower()
    if "not applicable" in t:
        return "not_applicable"
    # you can add more heuristics here later if needed
    return "has_content"

item6_summary = {"empty": 0, "not_applicable": 0, "has_content": 0}
examples = {"not_applicable": [], "has_content": []}

for file in os.listdir(extract_path)[:20]:  # adjust slice as needed
    if not file.endswith(".json"):
        continue
    with open(os.path.join(extract_path, file), "r", encoding="utf-8") as f:
        entry = json.load(f)
    item6 = (entry.get("item_6") or "").strip()
    cls = classify_item6(item6)
    item6_summary[cls] += 1

    # Save a couple short examples of each type
    if cls in ["not_applicable", "has_content"] and len(examples[cls]) < 3:
        examples[cls].append({
            "file": file,
            "preview": item6[:300].replace("\n", " ")
        })

print("Item 6 classification summary:", item6_summary)
print("\nExamples where Item 6 is 'not_applicable':")
for ex in examples["not_applicable"]:
    print(ex["file"], "->", ex["preview"])

print("\nExamples where Item 6 'has_content':")
for ex in examples["has_content"]:
    print(ex["file"], "->", ex["preview"])
```
